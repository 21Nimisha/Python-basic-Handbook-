# Line Plot -

A line plot can be plotted using the plot method. It plots Y versus X as lines
and/or markers

```python
# Defining coordinates to be plotted on X and Y axes respectively


x = [1.3, 2.9, 3.1, 4.7, 5.6, 6.5, 7.4, 8.8, 9.2, 10]
y = [95, 42, 69, 11, 49, 32, 74, 62, 25, 32]
```

```python
# Plot lists 'x' and 'y'
plt.plot(x, y)

# Plot axes labels and show the plot
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

 The plots created uses the default line style and color. The optional parameter
 fmt in the plot method is a convenient way for defining basic formatting
 like color, marker, and line-style

 fmt = '[color][marker][line]'

```python
# Plot line with green color
plt.plot(y, 'g')

# Plot axes labels and show the plot
plt.xlabel('Index Values')
plt.ylabel('Elements in List Y')
plt.show()
```

```python
# Plot continuous green line with circle markers
plt.plot(y, 'go-')
# Plot axes labels and show the plot
plt.xlabel('Index Values')
plt.ylabel('Elements in List Y')
plt.show()

# g refers to the green color, o refers to circle markers
# and - refers to a continuous line to be plotted.
```

```python
# Plot continuous green line with asterisk markers
plt.plot(y, 'g*-')
# Plot axes labels and show the plot
plt.xlabel('Index Values')
plt.ylabel('Elements in List Y')
plt.show()
```

 If we are to plot line
and markers with different colors, we can use multiple plot methods to
achieve the same.

```python
# Plot list 'y'
plt.plot(y, 'g')
# Plot red circle markers
plt.plot(y, 'ro')  #I firstly plot the line with the default style and then attempt to plot
# markers with attributes r referring to red color and o referring to circle
# Plot axes labels and show the plot
plt.xlabel('Index Values')
plt.ylabel('Elements in List Y')
plt.show()
```

```python
# Define two lists
y = [95, 42, 69, 11, 49, 32, 74, 62, 25, 32]
y2 = [35, 52, 96, 77, 36, 66, 50, 12, 35, 63]
# Plot lists and show them
plt.plot(y , 'go-')
plt.plot(y2 , 'r*-')

plt.xlabel('index values')
plt.ylabel('Elements in List Y')
plt.show()
```

```python
plt.plot(y , 'go-' , y2 , 'r*-')
plt.xlabel('Index values')
plt.ylabel('Elements in List Y')
```

#  Scatter Plot -

Scatter plots are used to visualize the relationship between two different
data sets. 

plt.scatter generates scatter plot of y vs x with varying marker size
and/or color.

```python
# Creating a dictionary with three key-value pairs
dictionary = {'a': np.linspace(1, 100, 50),
            'c': np.random.randint(0, 50, 50),
             'd': np.abs(np.random.randn(50)) * 100}
```

```python
dictionary
```

```python
# Creating a new dictionary key-value pair
dictionary['b'] = dictionary['a'] * np.random.rand(50)
```

```python
dictionary
```

```python
#Plotting a scatter plot using argument 'data'
plt.scatter('a', 'b', c='c', s='d', data=dictionary)

# Labeling the plot and showing it
plt.xlabel('A Data Points')
plt.ylabel('B Data Points')
plt.show()

# Values in key a and b contain fifty random values to be plotted on a scatter plot.
# Key c contains fifty random integers and key d contains fifty positive floats
```

# Histogram Plots -

A histogram is a graphical representation of the distribution of data. It is a
kind of bar graph and a great tool to visualize the frequency distribution of data.

To construct a histogram, the first step is to bin the range of data values, divide the entire
range into a series of intervals and finally count how many values fall into each interval.

All the bins/groups go on X-axis, and Y-axis shows the frequency of each
bin/group.


The hist method also take bins as an optional argument.
If this argument is specified, bins will be created as per the specified
value, otherwise, it will create bins on its own.

```python
#Data values for creating a histogram
y = [95, 42, 69, 11, 49, 32, 74, 62, 25, 32]
```

```python
# Creating a histogram
plt.hist(y)
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating a histogram with specified bin
plt.hist(y, bins= 20)
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating an array
array = np.random.normal(0, 1, 10000)
```

```python
plt.hist(array)
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()

#the data distribution indeed resembles a normal distribution.
```

The color of a histogram can be changed using
the color argument.

```python
# Creating an array
array = np.random.normal(0, 1, 10000)
```

```python
# Creating a histogram and plotting it
plt.hist(array, color='purple', histtype='step')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

orientation argument - This argument takes either of two values:
horizontal or vertical. The default is vertical.

```python
# Creating a histogram and plotting it
plt.hist(array, color='teal', orientation='horizontal')
plt.xlabel('Frequency')
plt.ylabel('Bins')
plt.show()
```

# X-Ticks & Y-Ticks method use for avoiding Overlapping
values on the X-axis are all overlapped with each other.
This happens as the plot method did not find sufficient space for each date.

The xticks method along with the rotation argument is used to rotate the
values/tick names along the x-axis.

```python
plt.plot(close_prices)
plt.xticks(rotation=45, color='teal', size=12)
```

```python
# Creating a figure, setting its size and plotting close
# prices on it
fig = plt.figure(figsize=(10, 5))
plt.plot(close_prices, color='purple')

# Customizing the axes
plt.xticks(rotation=45, color='teal', size=12)
plt.yticks(rotation=45, color='teal', size=12)

# Setting axes labels
plt.xlabel('Dates', {'color': 'orange', 'fontsize':15})
plt.ylabel('Prices', {'color': 'orange', 'fontsize':15})
plt.show()
```

Two main components that are missing in the
above plot are title and legend, which can be provided using the methods
title and legends respectively.

```python
# Showing legends and setting the title of plot
plt.legend()
plt.title('AAPL Close Prices', color='purple', size=20)

# Adding the grid to the plot
plt.grid(True)
```

```python
# Creating a figure, setting its size and plotting close
# prices on it
fig = plt.figure(figsize=(10, 5))
plt.plot(close_prices, color='purple')

# Customizing the axes
plt.xticks(rotation=45, color='teal', size=12)
plt.yticks(rotation=45, color='teal', size=12)

# Setting axes labels
plt.xlabel('Dates', {'color': 'orange', 'fontsize':15})
plt.ylabel('Prices', {'color': 'orange', 'fontsize':15})

plt.legend()
plt.title('AAPL Close Prices', color='purple', size=20)

# Adding the grid to the plot
plt.grid(True)

plt.show()
```

# axhline method -

The axhline method allows us to add a horizontal line across the axis to
the plot.

For example, we might consider adding the mean value of close
prices to show the average price of a stock for the whole duration. It can be
added using axhline method.

```python
# Importing NumPy library
import numpy as np
```

```python
# Calculating the mean value of close prices
mean_price = np.mean(close_prices)
```

```python
# Plotting the horizontal line along with the close prices
plt.axhline(mean_price, color='r', linestyle='dashed')
```

# text method -

text method provided by matplotlib.pyplot module to plot text
